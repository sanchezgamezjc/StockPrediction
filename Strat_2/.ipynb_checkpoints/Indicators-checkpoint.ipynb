```python
import pandas_ta as pta
import yfinance as yf
import datetime as dt
import pandas as pd
import talib as ta
```

```python
sample_per = 10000
stock = 'EURGBP=X'

end = dt.date.today()- dt.timedelta(days=2)
start = end - dt.timedelta(days=(sample_per))
interval = '1d'
```

```python
dataframe = yf.download(stock, start=start, end=end, interval=interval)
```
